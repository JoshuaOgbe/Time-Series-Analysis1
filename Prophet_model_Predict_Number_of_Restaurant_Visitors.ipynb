```python
from fbprophet import Prophet
import pandas as pd

homedepot = pd.read_csv('merged_df.csv', parse_dates=['period'])
```

```python
homedepot = homedepot[:-1]
```

```python
prophetdataframe = pd.DataFrame()
prophetdataframe['ds'] = homedepot['period']
prophetdataframe['y'] = homedepot['value']
prophetdataframe
```

```python
train = prophetdataframe[(prophetdataframe['ds'] < '2017-06-01')]
train
```

```python
test = pd.DataFrame(prophetdataframe[(prophetdataframe['ds'] >= '2017-06-01')].iloc[:, 0])
test
```

```python
m = Prophet()
```

```python
m.fit(train)
```

```python
forecast = m.predict(test)
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```

```python
test_predict = forecast[['ds', 'yhat']] 
test_predict
```

```python
test_actual = pd.DataFrame(prophetdataframe[(prophetdataframe['ds'] >= '2017-06-01')])
test_actual
```

```python
error_table = test_actual.merge(test_predict, on = 'ds')[:-1]
error_table
```

```python
spe = 0.0000
for i in range(len(error_table)):
    spe = spe + ((error_table['y'][i] - error_table['yhat'][i])*100/error_table['y'][i])**2
mspe = spe/len(error_table)
mspe
```

```python
import matplotlib.pyplot
import matplotlib.dates

error_table["ds"] = pd.to_datetime(error_table["ds"])
error_table.plot(x="ds", y=["y", "yhat"])
```
